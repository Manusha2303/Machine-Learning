```python
for c in cols:
    sk=dfsonar.skew().round()
    print(sk)
```

```python
for c in cols:
    sk=dfsonar[c].skew().round()
    if sk==0 or sk==-0 :
        m=dfsonar[c].mean()
        s=dfsonar[c].std()
        lower=m-(3*s)
        upper=m+(3*s)
        dfsonar[c]=dfsonar[c].map(lambda x : lower if x<lower else upper if x>upper else x)
    else:
        q1=dfsonar[c].quantile(0.25)
        q3=dfsonar[c].quantile(0.75)
        iqr=q3-q1
        low=q1-(1.5*iqr)
        up=q3+(1.5*iqr)
        dfsonar[c]=dfsonar[c].map(lambda x : low if x<low else up if x>up else x)
        
```
