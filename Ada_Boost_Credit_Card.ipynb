```python
cols=['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']
```

```python
for c in cols:
    print(c)
```

```python
for c in cols:
    sk=dfcard[c].skew().round()
    print(c,"==>",sk)
```

```python
for c in cols:
    sk=dfcard[c].skew().round()
    if sk==0 or-0:
        m=dfcard[c].mean()
        s=dfcard[c].std()
        lower=m-(3*s)
        upper=m+(3*s)
        dfcard[c]=dfcard[c].map(lambda x: lower if x<lower else upper if x>upper else x)
    else:
       q1=dfcard[c].quantile(0.25)
       q3=dfcard[c].quantile(0.75)
       iqr=q3-q1
       low=q1-(1.5*iqr)
       high=q3+(1.5*iqr)
       dfcard[c]=dfcard[c].map(lambda x: low if x<low else high if x>high else x)
```
